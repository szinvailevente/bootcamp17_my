```python
cegstat = pd.read_csv(r'C:\Users\szinv\Documents\Git_files\bootcamp17\data\rovat_2.csv', dtype=object)
```

```python
def find_alatt(nev):
    return 'alatt' in nev.lower()
```

```python
doglott=cegstat['nev'].map(find_alatt)
```

```python
cegstat['doglott']=doglott
```

```python
cegstat_doglott=cegstat[cegstat['doglott']]
```
